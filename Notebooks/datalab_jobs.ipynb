## NYSE Jobs with string 'cardiovascular tech'

### method: 
- read csv file (25GB) of NYSE Job Listings in chunks
- check data shape and save to SQLlite database
- query for specific job roles: 'cardiovascular tech'

```python
import os.path
from os import path
import pandas as pd
from sqlalchemy import create_engine
import sqlite3
```

```python
input_file = '../data/temp_datalab_records_job_listings.csv'
test_file = pd.read_csv(input_file, nrows = 2000000)
test_file.to_csv(r'../data/test_datalab.csv')
test_file.shape
```

```python
data_header = pd.read_csv(input_file, nrows = 50)
df = pd.DataFrame(data_header)
df
```

```python
csv_database = create_engine('sqlite:///csv_database.db')
csv_db = sqlite3.connect("csv_database")
```

```python
# Load the CSV in chunks:
for c in pd.read_csv("../data/test_datalab.csv", chunksize=10000):
    # Append all rows to a new database table, which we name 'datalab_test5':
    c.to_sql("cardio_tech", csv_db, if_exists="append")
csv_db.close()
```

```python
con = sqlite3.connect('csv_database')
cursor = con.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
# con.close()
```

```python
cursor.execute("select * from cardio_tech")
results = cursor.fetchall()
print (len(results))
con.close()
```

```python
con = sqlite3.connect('csv_database')
cardio_df = pd.read_sql_query("SELECT * from cardio_tech WHERE title LIKE '%cardiovascular tech%';", con)
cardio_df
con.close()
```
