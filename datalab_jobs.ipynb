```python
import os.path
from os import path
import pandas as pd
from sqlalchemy import create_engine
import sqlite3
```

```python
input_file = 'temp_datalab_records_job_listings.csv'
data_header = pd.read_csv(input_file, nrows = 10)
```

```python
df = pd.DataFrame(data_header)
df
```

```python
csv_database = create_engine('sqlite:///csv_database.db')
```

```python
col_list = ["dataset_id", "title","category","description"]

chunksize = 25000

i = 0
j = 1

for df in pd.read_csv(input_file, chunksize=chunksize, usecols=col_list, iterator=True):
      df = df.rename(columns={c: c.replace(' ', '') for c in df.columns}) 
      df.index += j
      i+=1
      df.to_sql('table', csv_database, if_exists='append')
      j = df.index[-1] + 1
```
